```python
df = pd.concat([df_analyst,df_scientist,df_engineers],axis=0)
```

```python
print(df.shape)
print(df.isnull().sum())
df.dropna(inplace=True, axis=0)
```

```python
### CONCAT e DROPNA feito!!








```

```python
### LIMPAR COLUNA "Job Title" 

# separar Job Titles por posições e seniority
def job_title(title):
    if "data analyst" in title.lower():
        return "data analyst"
    elif ("engineer" in title.lower()) or ("engineers" in title.lower()) or ("machine learning" in title.lower()) or ("ml" in title.lower()):
        return "data engineer"
    elif ("business" in title.lower()) or ("bi" in title.lower()) or ("marketing" in title.lower() or ("business intelligence" in title.lower())):
        return "business analyst"
    elif ("scientist" in title.lower()) or ("master" in title.lower()) or ("science" in title.lower()) or ("manager" in title.lower()) or ("director" in title.lower()):
        return "data scientist"
    else:
        return "na"
            
def seniority(title):
    if ("sr" in title.lower()) or ("sr." in title.lower()) or ("senior" in title.lower()) or ("lead" in title.lower()) or ("principal" in title.lower()):
        return "senior"
    elif ("jr" in title.lower()) or ("jr." in title.lower()) or ("junior" in title.lower()) or ("intern" in title.lower()) or ("developer" in title.lower()):
        return "junior"
    else:
        return "na"
        

df["job_title"] = df["Job Title"].apply(job_title)
df["seniority"] = df["Job Title"].apply(seniority)
```

```python
# dropar coluna "Job Title"
df.drop(columns=["Job Title"],axis=1,inplace=True)

```

```python
### LIMPAR COLUNA "Salary Estimate"

# retirar salários "-1" 
df = df[df["Salary Estimate"] != "-1"]

# limpar texto (podemos usar RegEx)
clean_salary = df["Salary Estimate"].apply(lambda x: x.replace("K","").replace("$","").replace("Per Hour","").split("(")[0])

# dividir salários em 2 (min-max salary)
df["min_salary"] = clean_salary.apply(lambda x: int(x.split("-")[0]))
df["max_salary"] = clean_salary.apply(lambda x: int(x.split("-")[1]))



# avg salary
df["avg_salary"] = (df["min_salary"]+df["max_salary"])/2
```

```python
# dropar coluna "Salary Estimate"
df.drop(columns=["Salary Estimate"],axis=1,inplace=True)
```

```python
### LIMPAR COLUNA "Easy Apply"

### NOTEBOOK KAGGLE
df["Easy Apply"] = df["Easy Apply"].replace("-1",False).astype("bool")
df["easy_apply"] = df["Easy Apply"].astype(int)
```

```python
# dropar coluna "Easy Apply"
df.drop(columns=["Easy Apply"],axis=1,inplace=True)
```

```python
### LIMPAR COLUNA "Job Description" 


# queremos ver q skills existem para cada Job Description

# Python
df["python"] = df["Job Description"].apply(lambda x: 1 if "python" in x.lower() else 0)
# machine learning/ml
df["ml"] = df["Job Description"].apply(lambda x: 1 if ("ml" in x.lower() or "machine learning" in x.lower() or "Artificial Intelligence" in x.lower()) else 0)
# statistics
df["statistics"] = df["Job Description"].apply(lambda x: 1 if "statistics" in x.lower() else 0)
# Java/JavaScript
df["js"] = df["Job Description"].apply(lambda x: 1 if ("java" in x.lower() or "javascript" in x.lower()) else 0)
# Spark
df["spark"] = df["Job Description"].apply(lambda x: 1 if "spark" in x.lower() else 0)
# AWS
df["aws"] = df["Job Description"].apply(lambda x: 1 if "aws" in x.lower() else 0)
# Excel
df["excel"] = df["Job Description"].apply(lambda x: 1 if "excel" in x.lower() else 0)
# SQL
df["sql"] = df["Job Description"].apply(lambda x: 1 if ("sql" in x.lower() or "structured query language" in x.lower()) else 0)
# SPSS
df["spss"] = df["Job Description"].apply(lambda x: 1 if "spss" in x.lower() else 0)
# Power BI
df["power_bi"] = df["Job Description"].apply(lambda x: 1 if ("powerbi" in x.lower() or "power bi" in x.lower() or "power-bi" in x.lower()) else 0)
# Tableau
df["tableau"] = df["Job Description"].apply(lambda x: 1 if "tableau" in x.lower() else 0)
# TensorFlow
df["tensorflow"] = df["Job Description"].apply(lambda x: 1 if ("tensorflow" in x.lower() or "tensor-flow" in x.lower()) else 0)



# Job Description Length (ver se descrições de trabalho mais longas corresponde ao valor de salário)
df["desc_len"] = df["Job Description"].apply(lambda x: len(x))
```

```python
# dropar coluna "Job Description"
df.drop(columns=["Job Description"],axis=1,inplace=True)
```

```python
### LIMPAR COLUNA "Company Name"

# limpar texto (retirar "\nrating")
df["company_name"] = df.apply(lambda x: x["Company Name"] if x["Rating"]<0 else x["Company Name"][:-4],axis=1)
```

```python
# dropar coluna "Company Name"
df.drop(columns=["Company Name"],axis=1,inplace=True)
```

```python
# Limpar coluna "Location"
df["job_state"] = df["Location"].apply(lambda x: x.split(",")[1]).str.replace("Arapahoe","CO").str.replace("United Kingdom","NY")

# dropar coluna "Location"  -> EM BAIXO
```

```python
df.columns
```

```python
### LIMPAR coluna "Size"
df["Size"].replace({"1 to 50 employees":"small","51 to 200 employees":"small",
                    "201 to 500 employees":"medium","501 to 1000 employees":"medium",
                   "1001 to 5000 employees":"large","5001 to 10000 employees":"large","10000+ employees":"large",
                   "-1":"other","Unknown":"other"},inplace=True)
```

```python
# ver se job_state está na mm localização q a Headquarters (dummies)
df["same_state"] = df.apply(lambda x: 1 if x["Location"] == x["Headquarters"] else 0, axis=1)
```

```python
# dropar coluna "Location"
df.drop(columns=["Location"],axis=1,inplace=True)
```

```python
### Limpar coluna "Founded" (idade da empresa)

df["age"] = df["Founded"].apply(lambda x: x if x<1 else 2020-x)
```

```python
# dropar coluna "Founded"
df.drop(columns=["Founded"],axis=1,inplace=True)
```

```python
# dropar coluna "Revenue"
df.drop(columns=["Revenue"],axis=1,inplace=True)
```

```python
### LIMPAR coluna "Competitors"

# nº competidores (-1: no competitors listed)
# split: retorna lista
df["num_competitors"] = df["Competitors"].apply(lambda x: len(x.split(",")) if x != "-1" else 0)
```

```python
# dropar coluna "Competitors"
df.drop(columns=["Competitors"],axis=1,inplace=True)
```
