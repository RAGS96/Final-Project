```python
### LIMPAR aqui as colunas que faltam 


# LIMPAR coluna "Type of ownership"
ordered_types_of_ownership = df["Type of ownership"].value_counts().reset_index()
df["Type of ownership"].replace(to_replace=list(ordered_types_of_ownership[2:]["index"].values), value="other", inplace=True)
df["type_of_ownership"] = df["Type of ownership"]


### Limpar coluna "Industry"
ordered_industry = df["Industry"].value_counts().reset_index()
df["Industry"].replace(to_replace=list(ordered_industry[7:]["index"].values), value="other", inplace=True)
df["industry"] = df["Industry"]


### LIMPAR coluna "Sector"
ordered_sector = df["Sector"].value_counts().reset_index()
df["Sector"].replace(to_replace=list(ordered_sector[7:]["index"].values), value="other", inplace=True)
df["sector"] = df["Sector"]


### LIMPAR coluna "job_state"
ordered_job_state = df["job_state"].value_counts().reset_index()
df["job_state"].replace(to_replace=list(ordered_job_state[6:]["index"].values), value="other", inplace=True)
```

```python
### dropar aqui as colunas limpadas


# dropar coluna "Type of ownership"
df.drop(columns=["Type of ownership"],axis=1,inplace=True)

# dropar coluna "Industry"
df.drop(columns=["Industry"],axis=1,inplace=True)

# dropar coluna "Sector"
df.drop(columns=["Sector"],axis=1,inplace=True)
```
